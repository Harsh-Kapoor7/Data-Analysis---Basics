```python
df = pd.read_csv('D:\Data_Analysis/data.csv')
df
```

```python
df.info()
```

```python
pd.to_numeric(df['Salary'])
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean') #initialising imputer

imputer.fit(df.iloc[:, 2:3].values) #doing calculation

df.iloc[:, 2:3] = imputer.transform(df.iloc[:, 2:3].values) #returning the thing after calculation
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy = 'most_frequent')
imputer.fit(df.iloc[:, 2:3].values)
df.iloc[:, 2:3] = imputer.transform(df.iloc[:, 2:3].values)
```

```python
df
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(df.iloc[:, :1].values)
df.iloc[:, :1] = imputer.transform(df.iloc[:, :1].values)
```
