```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean') # initialising imputer

imputer.fit(df.iloc[:,1:3].values) #doing calculation

df.iloc[:,1:3] = imputer.transform(df.iloc[:,1:3].values) #returning the thing after calculation 
```
