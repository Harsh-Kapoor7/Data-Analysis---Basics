```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
dict = {}

for name in df['Country'].unique():
    ct = 0
    for i in df['Country']:
        if i == name:
            ct += 1
    dict[name] = ct
dict
```

```python
for name in df['Purchased'].unique():
    ct = 0
    for i in df['Purchased']:
        if i == name:
            ct += 1
    print(name, ct)
```

## Removing Null Values for COUNTRY

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') #initialising imputer

imputer.fit(df.iloc[:,:1]) #doing manipulation

df.iloc[:,:1] = imputer.transform(df.iloc[:,:1]) #writing the thing after manipulation
```

```python
df
```

## Removing Null Values for PURCHASED

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') # initialising imputer

imputer.fit(df.iloc[:,3:]) #doing manipulation

df.iloc[:,3:] = imputer.transform(df.iloc[:,3:]) #wriring the things after manipulation
```
